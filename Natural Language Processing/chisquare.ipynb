```python
import nltk
from nltk.collocations import *
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import string
import math
```

```python
text=open('file.txt','r').read().lower()

text=text.translate(str.maketrans('','',string.punctuation))

text=word_tokenize(text)

sw=set(stopwords.words('english'))

text1=[]
for i in text:
 if i in sw:
  pass
 else:
  text1.append(i)

n=len(text1)
x=n
n=((n*(n-1))/2)

def chi_test(w1,w2,c1,c2,c12):
 chi=0
 r=n-c12
 t=n
 p_w1=(c1)/t
 p_w2=(c2)/t
 exp_mean=p_w1*p_w2*t
 ob_mean=c12/t
 diff=(ob_mean-exp_mean)
 v=pow(diff,2)
 v=v/exp_mean
 chi+=v
 
 p_w1=(x-c1)/t
 p_w2=(c2)/t
 exp_mean=p_w1*p_w2*t
 ob_mean=(c2-c12)/t
 diff=ob_mean-exp_mean
 v=pow(diff,2)
 v=v/exp_mean
 chi+=v
 
 p_w1=(c1)/t
 p_w2=(x-c2)/t
 exp_mean=p_w1*p_w2*t
 ob_mean=(c1-c12)/t
 diff=ob_mean-exp_mean
 v=pow(diff,2)
 v=v/exp_mean
 chi+=v
 
 print(chi)

 p_w1=(x-c1)/t
 p_w2=(x-c2)/t
 exp_mean=p_w1*p_w2*t
 ob_mean=r/t
 diff=ob_mean-exp_mean
 print(ob_mean," ",exp_mean)
 print('ff',diff)
 v=pow(diff,2)
 v=v/exp_mean
 print(v)
 chi+=v
 
 d={}
 d['chi-value']=chi
 d['word1']=w1
 d['word2']=w2
 return d

chitest=[]
for i in range(len(text1)-1):
 w1=text1[i]
 w2=text1[i+1]
 w12=w1+" "+w2
 c1=0; c2=0; c12=0;
 for w in range(len(text1)-1):
  if text1[w]==w1:
   c1+=1
   if text1[w+1]==w2:
    c12+=1
  if text1[w]==w2:
   c2+=1
 chitest.append(chi_test(w1,w2,c1,c2,c12))

c=0.58

print(chitest)

for w in chitest:
 if w['chi-value']>c:
  print('\n',w['word1'],' and ',w['word2'],' are collocations','\n')
 else: 
  print(w['word1'],' and ',w['word2'],' are not collocations')
    
```
